Twitter is one of the most popular social media apps where people are free to share their opinions on any topic. There are so many tweets recorded about Pfizer vaccine awareness that can be used to analyze the sentiments of people about the Pfizer vaccine.

The dataset that I am using for the task of Pfizer vaccine sentiment analysis is downloaded from Kaggle, which was initially collected from Twitter when people were sharing their opinions about the Pfizer vaccine. Let‚Äôs start the task of Pfizer vaccine sentiment analysis by importing the necessary Python libraries and the dataset:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

data = pd.read_csv(r"C:\Users\SHREE\Downloads\Python CODES\Pfizer Vaccine Sentiment Analysis using Python\vaccination_tweets.csv")
print(data.head())
```

So this dataset is quite large, let‚Äôs have a look at whether it contains any null values or not:

```python
data.isnull().sum()
```

Although these null values will not affect the task of sentiment analysis, to keep things simple, I will drop the rows containing null values as the dataset is already large:

```python
data = data.dropna()
```

Now let‚Äôs have a look at the descriptive statistics of this dataset:

```python
print(data.describe())
```

The text column is the most important feature in this dataset as it contains the opinions of the users of Twitter about the Pfizer vaccine. But the text column needs to be prepared as it contains many special symbols and language errors. Below is how we can clean the text column:

```python
import nltk
import re
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
data["text"] = data["text"].apply(clean)
```

Now, let‚Äôs have a look at the word cloud of the text column. A word cloud is a data visualisation technique that shows the most used words in large font and the less used words in small font. Here is how you can visualize the word cloud of the text column:

```python
text = " ".join(i for i in data.text)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

Now let‚Äôs have a look at the word cloud of the hashtags column, which can show what kind of tags was trending when people were sharing their opinions about the Pfizer vaccine:

```python
text = " ".join(i for i in data.hashtags)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

The ‚Äúuser_verified‚Äù column in the dataset shows whether the users who have shared their opinions are verified by Twitter or not. A verified user on Twitter is a public figure or a celebrity. So let‚Äôs have a look at how many users were verified who shared their opinions about the Pfizer vaccine:

```python
data["user_verified"].value_counts()
```

In the above output, False shows the count of unverified users and True shows the count of verified users. Now let‚Äôs move to the task of sentiment analysis of the Pfizer vaccine. Here I will add three more columns in this dataset as Positive, Negative, and Neutral by calculating the sentiment scores of the text column:

```python
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["text"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["text"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["text"]]
data = data[["text", "Positive", "Negative", "Neutral"]]
print(data.head())
```

Now let‚Äôs calculate how most of the people felt about the Pfizer vaccine:

```python
x = sum(data["Positive"])
y = sum(data["Negative"])
z = sum(data["Neutral"])

def sentiment_score(a, b, c):
    if (a>b) and (a>c):
        print("Positive üòä ")
    elif (b>a) and (b>c):
        print("Negative üò† ")
    else:
        print("Neutral üôÇ ")
sentiment_score(x, y, z)
```

So most of the opinions of the users were Neutral, let‚Äôs have a look at the total of each sentiment score before making any conclusion:

```python
print("Positive: ", x)
print("Negative: ", y)
print("Neutral: ", z)
```

The total of positive and negative is very less than Neutral, so we can say that the discussion of the Twitter users was about the awareness of the Pfizer vaccine rather than sharing its benefits or drawbacks.
